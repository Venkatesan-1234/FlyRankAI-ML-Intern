## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

This project functions strictly as an observational, decision-support tool. It ranks content candidates by identifying statistical vulnerabilities and visibility gaps based on observed historical metrics.

What I can claim: I can claim that certain observable patterns (such as a high impression-to-click gap or an extended period without a content update) correlate with declining performance, and that our model can prioritize these patterns more effectively than a fixed rule baseline.

What I cannot claim: I cannot claim that our ranking proves Google algorithm ranking factors, nor can I claim that performing a content refresh will cause a traffic recovery. Proving causality requires isolated, controlled experimentation, whereas this model handles prioritization based on directional trends under limited operational capacity.
